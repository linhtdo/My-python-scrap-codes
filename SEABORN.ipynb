```python
# PAIRPLOT: plotting pairwise relationship across an entire dataframe
sns.pairplot(tips, hue='sex', palette='coolwarm')

#palette change charts to specific color
```

# Categorical Plots

```python
import seaborn as sns
%matplotlib inline
import numpy as np

tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

Average total bill for male vs female

Use estimator argument to change the type of statistical function to estimate within each category

```python
# COUNTPLOT: counting the occurences of categories
sns.countplot(x= 'sex',data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data= tips, hue='smoker')
# Hue aregument: segmenting guess (smoker vs non-smoker)
```

```python
# VILOLINPLOT
sns.violinplot(x='day', y='total_bill', data= tips)
```

Violin plot demonstrates the distribution of all the data point, as well as the kde for each category

```python
sns.violinplot(x='day', y='total_bill', data= tips, hue='sex',split=True)
```

##### Using hue argument to segment the data based on gender, and split to make it look super neat

```python
#STRIPPLOT
sns.stripplot(x='day', y='total_bill',data= tips,jitter=True)
```

```python
sns.stripplot(x='day', y='total_bill',data= tips,jitter=True, hue='sex', split=True)
```

```python
#SWARMPLOT: Like stripplot, but the points are adjusted so that they dont overlap
sns.swarmplot(x='day', y= 'total_bill', data= tips)
```

```python
Swarm plot can be combined with violin plot
```

```python
sns.violinplot(x='day', y='total_bill', data= tips)
sns.swarmplot(x='day', y= 'total_bill', data= tips, color= 'yellow')
```

```python
# FACTOR PLOT: the most general plot of them all
sns.factorplot(x='day', y='total_bill', data= tips, kind='bar')
```

# Matrix Plot (aka Heat Map)

```python
import seaborn as sns
%matplotlib inline
import numpy as np

tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

```python
# CORRELATION DATA:
tc = tips.corr()
tc
```

```python
sns.heatmap(tc, annot= True)

# Can use cmap= 'coolwarm' to specify map's colors
```

```python
# PIVOT TABLE DATA:
fv = flights.pivot_table(index='month', columns='year', values= 'passengers')
fv
```

```python
sns.heatmap(fv)
```

People travel more during the summer

```python
sns.heatmap(fv, cmap= 'magma', linecolor= 'white', linewidths=1)
```

```python
sns.clustermap(fv)
```

Cluster Map: Clustering rows and columns together based off of their similarities

```python
sns.clustermap(fv, cmap= 'coolwarm', standard_scale= 1)
```

More coming with machine learning clustering algorithms

# Grids
More about subplots

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
# Grids give a lot more control over graphing

g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)

g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot) #cool customizations
```

```python
# FACET GRID

tips = sns.load_dataset('tips')
tips.head()
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
# then we can start mapping 
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(sns.distplot, 'total_bill')
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(plt.scatter, 'total_bill', 'tip')
# Since scatter requires more than 1 argument, need to pass additiional argument in
```

# REGRESSION PLOTS

```python
import seaborn as sns
%matplotlib inline

tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x ='total_bill', y='tip',data=tips, hue='sex', markers=['o', 'v'],
          scatter_kws={'s':100})
# scatter_kws used to adjust marker sizes
```

```python
sns.lmplot(x ='total_bill', y='tip',data=tips, col='sex', row='time') 
#Separating genders by columns and time by rows
```

```python
sns.lmplot(x ='total_bill', y='tip',data=tips, col='day', hue='sex',
          aspect =0.6, size = 8) 
#aspect and size arguments are used to adjust graph ratios and sizes
```
